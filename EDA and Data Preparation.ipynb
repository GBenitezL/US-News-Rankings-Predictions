```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 30)
```

```python
df1 = pd.read_csv('US News Overall 2018-2019.csv')
df1['Year'] = 2019
df2 = pd.read_csv('US News Overall 2019-2020.csv')
df2['Year'] = 2020
df3 = pd.read_csv('US News Overall 2020-2021.csv')
df3['Year'] = 2021 
df4 = pd.read_csv('US News Overall 2021-2022.csv')
df4['Year'] = 2022
```

```python
rankings = pd.concat([df1, df2, df3, df4])
rankings.head(10)
```

```python
plt.figure(figsize=(10,6))
sns.displot(rankings['Score'], bins = 10)
plt.title('Puntuaciones por Rangos')
plt.ylabel('Nmero de Universidades')
plt.xlabel('Rangos de Puntuacin')
plt.show()
```

```python
plt.figure(figsize=(10,6))
promedios = rankings.groupby('Region')['Score'].mean()
sns.barplot(x = promedios.keys(), y = promedios)
plt.xticks(rotation=90)
plt.title('Puntuacin Promedio por Regin')
plt.ylabel('Score')
plt.show()
```

```python
excluded = ['Name',	'Country',	'Region', 'RANK', 'Year']
for column in rankings.columns:
  if column not in excluded:
   rankings[column] = (rankings[column] - rankings[column].min()) / (rankings[column].max() - rankings[column].min())
rankings.describe()
```

```python
cercanas = pd.DataFrame()
cercanas['Name'] = unis
cercanas
```

```python
score_2019 = df1[['Name', 'Score']]
score_2020 = df2[['Name', 'Score']]
score_2021 = df3[['Name', 'Score']]
score_2022 = df4[['Name', 'Score']]
```

```python
tmp1 = pd.merge(cercanas, score_2019, on='Name', how='left')
tmp2 = pd.merge(tmp1, score_2020, on='Name', how='left')
tmp3 = pd.merge(tmp2, score_2021, on='Name', how='left')
cercanas = pd.merge(tmp3, score_2022, on='Name', how='left')
```

```python
cercanas.columns = ['Name', '2018-2019', '2019-2020', '2020-2021', '2021-2022']
cercanas
```
